```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = [(3, 1, 3), (7, 1, 2), (5, 1, 4), (7, 1, 2), (8, 2, 5),
     (4, 2, 4), (6, 4, 5), (8, 2, 5)]
g_s = sorted(g, key=lambda x: x[0])
vertex = set()
used_vertex = {}     
edge = []    
for r in g_s:
    if r[1] not in vertex or r[2] not in vertex:  
        if r[1] not in vertex and r[2] not in vertex: 
            used_vertex[r[1]] = [r[1], r[2]]          
            used_vertex[r[2]] = used_vertex[r[1]]               
        else:                         
            if not used_vertex.get(r[1]):            
                used_vertex[r[2]].append(r[1])        # добавляем в список первую вершину
                used_vertex[r[1]] = used_vertex[r[2]]           
            else:
                used_vertex[r[1]].append(r[2])        
                used_vertex[r[2]] = used_vertex[r[1]]
        edge.append(r)             # добавляем ребро в остов
        vertex.add(r[1])             # добавляем вершины в множество vertex
        vertex.add(r[2])
```

```python
for r in g_s:    
    if r[2] not in used_vertex[r[1]]:     
        edge.append(r)             # добавляем ребро в остов
        gr1 = used_vertex[r[1]]
        used_vertex[r[1]] += used_vertex[r[2]]      # объединем списки двух групп вершин
        used_vertex[r[2]] += gr1

print(edge)
```
