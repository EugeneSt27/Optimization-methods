```python
import numpy as np
import pandas as pd 
from scipy.optimize import linear_sum_assignment
import matplotlib.pyplot as plt
```

# Задание №1

# Подготовка данных

```python
data = pd.read_excel('case2.xlsx', header = 1)
pd.set_option('display.max_columns', None)
data
```

Каждому дистрибьютору нужно назначить 7 оптовиков, для этого продублируем каждую строку 7 раз и удалим лишние колонки

```python
newdf = pd.DataFrame(np.repeat(data.values, 7, axis=0))
costs = newdf[range(1,29)]
costs
```

```python
costs.columns = range(0,28)
costs
```

# Решение задачи

```python
row_ind, col_ind = linear_sum_assignment(costs)
print(row_ind, col_ind)
```

```python
numbers = []
for i in range(0,28):
    c = costs[col_ind[i]][row_ind[i]]
    numbers.append(c)
print(f'Total cost = {sum(numbers)}')
```

# Визуализация решения

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.axis('off')
table = ax.table(cellText=newdf.values, loc='center', rowLabels=newdf.index, colLabels = newdf.columns )

# красим ячейки
for i in range(len(row_ind)):
    table.get_celld()[(row_ind[i]+1, col_ind[i]+1)].set_facecolor('yellow')

plt.show()
```

чтобы наглядно представить результат на исходной таблице, переназначим индексы строк

```python
r = row_ind.tolist() # r row_ind всегда будет одинаковый, поэтому в дальнейшем не будем создавать новые r
for i in range(0,28):
    if r[i] < 7:
        r[i] = 0
    elif r[i] < 14:
        r[i] = 1
    elif r[i] < 21:
        r[i] = 2
    else:
        r[i] = 3
    i+=1
```

```python
fig, ax = plt.subplots(figsize=(12,2))
ax.axis('off')
table = ax.table(cellText=data.values, loc='center', colLabels = data.columns )

# красим ячейки
for i in range(len(row_ind)):
    table.get_celld()[(r[i]+1, col_ind[i]+1)].set_facecolor('yellow')
plt.title(f'Total cost = {sum(numbers)}', fontsize= 15)
plt.show()
```

# Задание №2

Транспортный  отдел  просит  не  назначать  оптовику  данный  центр,  если 
время доставки превышает 10 часов.   
Можно ли это обеспечить?   
Как при этом изменится суммарное время доставки?  

Чтобы "поставить запрет" на использование вариантов, которые превышают 10 часов, увеличим значения всех соответствующих ячеек в 10 раз

```python
costs1 = costs.copy()
for c in costs1.columns:
    costs1[c] = costs1[c].apply(lambda x: x*1000 if x >10 else x)
costs1
```

```python
row_ind1, col_ind1 = linear_sum_assignment(costs1)
print(row_ind1, col_ind1)
```

```python
numbers1 = []
for i in range(0,28):
    c = costs1[col_ind1[i]][row_ind1[i]]
    numbers1.append(c)
print(f'Total cost = {sum(numbers1)}')
  
```

```python
fig, ax = plt.subplots(figsize=(12,2))
ax.axis('off')
table = ax.table(cellText=data.values, loc='center', colLabels = data.columns )

# красим ячейки
for i in range(len(row_ind)):
    table.get_celld()[(r[i]+1, col_ind1[i]+1)].set_facecolor('yellow')
plt.title(f'Total cost = {sum(numbers1)}', fontsize= 15)
plt.show()
```

### Вывод

Можно обеспечить время доставки, не превышающее 10 часов   
В таком случае, суммарное время доставки увеличится на одну единицу
