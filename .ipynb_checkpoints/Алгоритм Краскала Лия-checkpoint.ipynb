```python
a = [[1, 2, 22], [1, 9, 27], [1, 8, 23], [2, 3, 29], 
     [3, 4, 29], [3, 9, 15], [3, 10, 20], [4, 5, 28], 
     [5, 6, 27], [5, 10, 31], [6, 7, 26], [6, 10, 30],
     [7, 8, 24], [7, 9, 19], [8, 9, 18], [9, 10, 14]]
a = sorted(a, key = lambda x: x[2])
print(a)
```

```python
def f1(a):
  i = a[0]
  Ves = []
  Ves.append(i[2])
  #i[0:2] - вершины
  V = []#вершины выбранные красные
  V.extend(i[0:2])
  #i[0] - 1 вершина, i[1] - 2 вершина
  W2 = []
  W2.append(tuple((i[0], i[1])))
  L = [] #длина ребер изолир (син)
  K = [] # вершины изолир 
  r = [] # ребра
  for i in a[1::]:
    if i[0] not in V or i[1] not in V: #Проверка, что нет циклa
      if i[0] not in V and i[1] not in V: 
          K.extend(i[0:2])
          r.append(tuple((i[0], i[1])))
          L.append(i[2])
      elif i[0] not in V:
        if i[0] in K and (i[1] not in K) and (i[1] not in V):
          K.append(i[1])
          L.append(i[2])
          r.append(tuple((i[0], i[1])))
        elif i[1] in V and i[0] not in K:
          V.append(i[0])
          Ves.append(i[2])
          W2.append(tuple((i[0], i[1])))
        elif i[0] in K and i[1] in V:
          r.append(tuple((i[0], i[1])))
          L.append(i[2])
          W2.extend(r)
          V.extend(K)
          Ves.extend(L)
      elif i[1] not in V:
        if i[1] in K and i[0] not in K:
          K.append(i[0])
          L.append(i[2])
          r.append(tuple((i[0], i[1])))
        elif i[1] in K and i[0] in V:
          W2.extend(r)
          W2.append(tuple((i[0], i[1])))
          V.extend(K)
          Ves.expend(L)
        else:
          W2.append(tuple((i[0], i[1])))
          V.append(i[1])                    
          Ves.append(i[2])
  return (W2, Ves, V)
```

```python
Kraskal = f1(a)
W2 = Kraskal[0]
print('Ребра', *Kraskal[0])
print('Остов минимальный =', sum(Kraskal[1]))
```

```python
G = nx.Graph()
G.add_weighted_edges_from(a)
```

```python
G = nx.Graph()
G.add_weighted_edges_from(a)
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw(G, pos, with_labels = True)
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
nx.draw_networkx_edges(G, pos, edgelist = W2, edge_color='pink')
plt.tight_layout()
plt.show()
```
