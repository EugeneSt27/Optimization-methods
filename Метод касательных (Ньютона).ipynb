```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Метод секущих

```python
def func(x):
    return x**4 + 7*x**2 +2*x +1
```

```python
def dfunc(x):
    return 4*x**3 + 14*x +2
```

```python
def d2func(x):
    return 12*x**2 + 14
```

```python
a = -3
b = 2
eps = 0.1
k = 1
z = b - dfunc(b)/d2func(b)
```

```python
x_list = []
f_list = []
while abs(dfunc(z)) > eps:
    z = b - dfunc(b)/d2func(b)
    if dfunc(z) < 0:
        a = z
    else:
        b = z
    x_opt = z
    f_opt = func(x_opt)
    print(f'На {k} итерации значение х равно {round(x_opt,3)}, f(x) = {round(f_opt,4)}')
    x_list.append(x_opt)
    f_list.append(f_opt)
    k+=1
print(f'Решение с точностью {round(abs(dfunc(z)),10)} найдено на {k-1} итерации , x* = {round(x_opt,4)}, f(x*) = {round(f_opt,4)} ')
```

# График функции

```python
plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(-5, 5, 100) 
y = x**4 + 7*x**2 +2*x +1
print(x_list, f_list,sep='\n')
plt.ylim([0,20])
plt.xlim([-1.5,1.5])
plt.plot(x, y, 'blue')
plt.scatter(x_list, f_list, c = 'red')
plt.show()
```
