# Задание №4

```python
def triangle(a,b,c,x):
    if (x>=0) and (x<=(c-a)/(b-a)):
        return ((x*(b-a)*(c-a))**0.5)+a
    elif (x>1-(b-c)/(b-a)) and (x<=1):
        return b-((1-x)*(b-a)*(b-c))**0.5
```

```python
a = 1
b = 4
c = 2.5
t_list = []
unilist = np.random.uniform(size = 50000)
for i in unilist:
    x = triangle(a,b,c,i)
    t_list.append(x)
```

```python
Mx = np.mean(t_list)
Dx = np.var(t_list)
print(f'Мат.ожидание составит {round(Mx,3)}, Дисперсия - {round(Dx, 3)}')
```

```python
sns.histplot(data = t_list, binwidth = 0.05, kde = True)
```

1) Преобразование Бокса-Мюллера

```python
import math
```

```python
def box_muller(n):
    x = [[np.random.uniform(), np.random.uniform()] for i in range(n)]
    u1 = [np.cos(2*math.pi*x[i][0])*math.sqrt(-2*math.log(x[i][1])) for i in range(n)]
    u2 = [math.sin(2*math.pi*x[i][0])*math.sqrt(-2*math.log(x[i][1])) for i in range(n)]
    u1.extend(u2);
    
    sns.histplot(u1, color='green', kde=True)
    return u1
```

```python
d = box_muller(10000)
```
