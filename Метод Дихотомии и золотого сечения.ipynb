```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Метод дихотомии

```python
def func(x):
    return x**4 + 7*x**2 +2*x +1
```

```python
a = -3
b = 2
eps = 0.1
k = 1
delta = 0.05 
epsk = (b-a)/2
```

```python
x_list = []
f_list = []
while epsk > eps:
    x1 = (a+b-delta)/2
    x2 = (a+b+delta)/2
    if func(x1)<= func(x2):
        b = x2
    else:
        a = x1
    epsk  = (b-a)/2
    x_opt = (b+a)/2
    f_opt = func(x_opt)
    print(f'На {k} итерации значение х равно {round(x_opt,3)}, f(x) = {round(f_opt,4)}')
    x_list.append(x_opt)
    f_list.append(f_opt)
    k+=1
print(f'Решение с точностью {round(epsk,3)} найдено на {k-1} итерации , x* = {round(x_opt,3)}, f(x*) = {round(f_opt,5)} ')
```

```python
plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(-1, 1, 100) 
y = x**4 + 7*x**2 +2*x +1
print(x_list, f_list,sep='\n')
plt.ylim([0,10])
plt.plot(x, y, 'blue')
plt.scatter(x_list, f_list, c = 'red')
plt.show()
```

# Метод золотого сечения

```python
a = -3
b = 2
eps = 0.1
T = (3-5**1/2)/2 
k = 1
l = b-a
```

```python
while l > eps:
    x1 = a+T*(b-a)
    x2 = a+b-x1
    if func(x1)<= func(x2):
        b = x2
        x2 = x1
        x1 = a+b-x2
    else:
        a = x1
        x1 = x2
        x2 = a+b-x1
    l = b-a
    x_opt = (b+a)/2
    f_opt = func(x_opt) 
    print(f'На {k} итерации значение х равно {round(x_opt,3)}, f(x) = {round(f_opt,4)}')
    x_list.append(x_opt)
    f_list.append(f_opt)
    k+=1
print(f'Решение с точностью {round(l,3)} найдено на {k-1} итерации , x* = {round(x_opt,3)}, f(x*) = {round(f_opt,5)} ')
```

```python
plt.xlabel('x')
plt.ylabel('y')
x = np.linspace(-1, 1, 100) 
y = x**4 + 7*x**2 +2*x +1
#print(x_list, f_list,sep='\n')
plt.ylim([0,10])
plt.xlim([-1,1])
plt.plot(x, y, 'blue')
plt.scatter(x_list, f_list, c = 'red', s = 20)
plt.show()
```
